## Librarys and Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import shap

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
```

```python
# as finaly decided in 01_analyze_data.ipynb

df = pd.read_csv(r'..\..\data\Healthcare_Diabetes_Dataset\Healthcare-Diabetes.csv')
df = df.query('Glucose != 0 and BloodPressure != 0 and SkinThickness != 0 and BMI != 0')
df = df.drop_duplicates(subset=df.columns.difference(['Id']), keep='first')
df = df.drop(columns=['Pregnancies'])
df.loc[:,'Outcome_dif'] = np.select([(df['Outcome'] == 1) & (df['Insulin'] > 20), (df['Outcome'] == 1) & (df['Insulin'] <= 20)], ['type_II', 'type_I'], default='healthy')
```

## Random Forest including Insulin as Feature
Since I used insulin to deterministically categorize healthy, type 1 diabetes and type 2 diabetes, it would not be surprising that a classifier would "find" this rule, even though there are also individuals with high insulin in the data set who are labeled as healthy based on the original outcom-variable.

### train the model (insulin as predictor included)

```python
# split the features (X) and targets (y) int to differentiated by Target ((healhty, typ_I, type_II) )
# reason: get an equal 80:20 split across all subsets (healhty, typ_I, type_II) 

X_healthy = df[df['Outcome_dif'] == 'healthy'].drop(['Outcome', 'Id', 'Outcome_dif'], axis=1)
y_healthy = df[df['Outcome_dif'] == 'healthy']['Outcome_dif']
X_type_I = df[df['Outcome_dif'] == 'type_I'].drop(['Outcome', 'Id', 'Outcome_dif'], axis=1)
y_type_I = df[df['Outcome_dif'] == 'type_I']['Outcome_dif']
X_type_II = df[df['Outcome_dif'] == 'type_II'].drop(['Outcome', 'Id', 'Outcome_dif'], axis=1)
y_type_II = df[df['Outcome_dif'] == 'type_II']['Outcome_dif']
```

```python
# split each of the subsets (healhty, typ_I, type_II) into test and traingings data
X_train_healthy, X_test_healthy, y_train_healthy, y_test_healthy = train_test_split(X_healthy, y_healthy, test_size=0.2, random_state=123)
X_train_type_I, X_test_type_I, y_train_type_I, y_test_type_I = train_test_split(X_type_I, y_type_I, test_size=0.2, random_state=123)
X_train_type_II, X_test_type_II, y_train_type_II, y_test_type_II = train_test_split(X_type_II, y_type_II, test_size=0.2, random_state=123)

# concatenate subsets (healhty, typ_I, type_II) of the trainings- and testdate
X_train = pd.concat([X_train_healthy, X_train_type_I, X_train_type_II])
X_test = pd.concat([X_test_healthy, X_test_type_I, X_test_type_II])
y_train = pd.concat([y_train_healthy, y_train_type_I, y_train_type_II])
y_test = pd.concat([y_test_healthy, y_test_type_I, y_test_type_II])
```

```python
#Initialize and train random forest model
rf_model = RandomForestClassifier(n_estimators=100, random_state=123)
rf_model.fit(X_train, y_train)
```

### evaluate the model quality (insulin as predictor included)

```python
# prediction
y_pred = rf_model.predict(X_test)

# model evaluation
print(classification_report(y_test, y_pred))
```

The overall accuracy of the model is 0.76, indicating that 76% of the predictions were correct.

<b>Healthy: </b>The model identifies healthy cases with a precision of 0.81 and a recall of 0.83. The F1-score of 0.82 shows a good balance between precision and recall. <br>

<b>Type_I: </b>The model struggles to recognize type_I cases, as indicated by the lower F1-score of 0.64. Given the small number of cases for type_I diabetes in the dataset, this result is acceptable. However, future analyses with larger datasets are necessary to improve the model's performance.<br>

<b>Type_II: </b>The model identifies type_II cases with a precision of 0.70 and a recall of 0.59. The F1-score of 0.64 is similar to that of type_I, indicating similar performance challenges.<br>

In summary, while the model performs reasonably well for healthy cases, it has difficulty accurately identifying both type_I and type_II diabetes cases. Taking into account that there were few data points available, pregnancies were not used as a feature, and the classifier was not optimized through feature transformation and hyperparameter tuning, the results are satisfactory.

Nevertheless, this highlights the need for larger datasets in future examinations to enhance the model's ability to distinguish between different types of diabetes effectively without considering pregnancies or other gender-specific variables

###  evaluate feature importance of the model via SHAP values (insulin as predictor included)

<b>Introduction to SHAP: </b><br>
SHAP values are a construct from game theory and assess how much a feature contributes to a prediction of a model. Positive SHAP values indicate that higher values of the feature tend to increase the prediction, while negative SHAP values suggest that higher values of the feature tend to decrease the prediction. They thus indicate the extent to which each feature influences the model's decision, with SHAP values close to zero suggesting that the feature is not significant for the prediction.

```python
# Calculate shaply values
explainer = shap.TreeExplainer(rf_model)
shap_values = explainer.shap_values(X_train)
```

```python
shap.summary_plot(shap_values, X_train, plot_type="bar", class_names=rf_model.classes_)
plt.show()
```

Insulin, glucose, and age are the most important predictors for the model to classify healthy individuals, Type 1 diabetes, and Type 2 diabetes. The insulin level is particularly relevant for classifying Type 1 and Type 2 diabetes, while glucose is crucial for distinguishing between healthy and non-healthy individuals.

DiabetesPedigreeFunction, BMI, and SkinThickness are approximately equally important for the model, but each provides only about one-third of the predictive power that insulin has for the model.

Since I want to exclude insulin as a predictor, I will examine the predictors glucose and age in the context of insulin more closely. I will also take a look at BMI, as Type 2 diabetes is associated with overweight.

```python
def plot_shap_dependence(model, shap_values, X_train, feature, interaction_feature):
    """
    Create a 1x3 subplot of SHAP dependence plots for a given feature and interaction feature.

    Parameters:
    model: The trained model.
    shap_values (list of np.ndarray): List of SHAP values for each class.
    X_train (pd.DataFrame): Training data.
    feature (str): The feature to plot.
    interaction_feature (str): The interaction feature to plot.
    """
    # Get the class labels from the model
    class_labels = model.classes_

    # Create a 1x3 subplot
    fig, axs = plt.subplots(1, 3, figsize=(18, 5))

    # Iterate over the classes to create subplots
    for i in range(3):
        shap.dependence_plot(feature, shap_values[i], X_train, interaction_index=interaction_feature, ax=axs[i], show=False)
        axs[i].set_title(class_labels[i])

    # Display the plot
    plt.tight_layout()
    plt.show()
```

```python
plot_shap_dependence(rf_model, shap_values, X_train, "Glucose", "Insulin")
```

The model classifies individuals as healthy when glucose levels are low and increasingly as not healthy when glucose levels rise above 120 mg/dL (140 mg/dL is considered prediabetes). Low glucose levels with low insulin indicate health. Accordingly, the model also recognizes that low insulin and low glucose argue against Type 1 diabetes, while high glucose with low insulin indicates Type 1 diabetes. Low insulin levels with high glucose are a more important predictor than high insulin levels with high glucose, which corresponds to the physiological mechanism behind Type 1 diabetes. For Type 2 diabetes, the model correctly identifies that higher insulin levels with high glucose strongly indicate Type 2 diabetes.

```python
plot_shap_dependence(rf_model, shap_values, X_train, "Age", "Insulin")
```

Regardless of insulin levels, the model evaluates younger age as an indicator of health and older age as indicative of non-health. Plausibly, the model's evaluation for Type 1 and Type 2 diabetes shows contrasting age assessments. Erroneously, young individuals with low insulin levels exhibit more negative SHAP values compared to young individuals with higher insulin levels (those with SHAP values close to ~0 are deemed unimportant for model decision-making).

```python
plot_shap_dependence(rf_model, shap_values, X_train, "BMI", "Insulin")
```

Largely regardless of insulin, the model considers higher BMI values to indicate non-health and lower BMI values to indicate health. Interestingly, the model tends to classify individuals as having Type 1 diabetes more strongly when BMI is low and insulin levels are low, and more strongly indicates Type 1 diabetes when BMI values are high while insulin levels are simultaneously low, compared to when insulin levels are normal to high. In the context of Type 2 diabetes, low insulin levels across all BMI levels have little significance. It also suggests that a lower BMI with normal to high insulin levels argues against Type 2 diabetes, while a higher BMI with high insulin levels argues for Type 2 diabetes. Overall, however, the picture is not as clear for Type 2 diabetes.
